# Exploratory Data Analysis

### Gender

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)

for bars in ax.containers:
    ax.bar_label(bars)
```

From above two graphs we can see that most of the buyers are Females and purchasing power of the female is more that the male.

### Age

```python
df.columns
```

```python
ax = sns.countplot(x = 'Age Group', hue = 'Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#total_amount Vs Age_Group
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age)
```

### State

```python
# total number of order from top 10 states:

state_ord = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)

sns.set(rc = {'figure.figsize':(18,5)})
sns.barplot(data = state_ord, x = 'State', y = 'Orders')
```

### Marital Status

```python
df.columns
```

```python
ax = sns.countplot(x = 'Marital_Status', data = df, hue = 'Gender')
sns.set(rc = {'figure.figsize':(5,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

### Occupation

```python
df.columns
```

```python
ax = sns.countplot(x = 'Occupation', data = df)
sns.set(rc = {'figure.figsize': (15, 5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occu = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(15,5)})
sns.barplot(data = sales_occu, x = 'Occupation', y = 'Amount')
```

### Product Category

```python
df.columns
```

```python
ax = sns.countplot(x = 'Product_Category', data = df)
sns.set(rc = {'figure.figsize':(35,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_product = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot(data = sales_product, x = 'Product_Category', y = 'Amount')
```

```python
From the above two graphs we can see that items of Clothing & Apparel sold more but the amount generate by the Food is more.
```
